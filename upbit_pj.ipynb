```python
import requests

def get_current_prices_api():
  portfolio_analysis = []
  total_value = 0

  portfolio = {
      "KRW-SHIB":174434595,
      "KRW-DOGE":6775
  }
  markets = ",".join(portfolio.keys())
  url = "https://api.upbit.com/v1/ticker"
  params = {"markets": markets}
  headers = {"accept": "application/json"}
  response = requests.get(url, headers=headers, params=params)

  for ticker in response.json():
    trade_price = ticker["trade_price"]
    coin_count = portfolio[ticker["market"]]
    total_price = trade_price * coin_count

    my_portfolio = {
        "coin" : ticker["market"],
        "count": coin_count, # 코인 보유개수
        "trade_price": trade_price,  #현재 거래가
        "total_price" : total_price # 내 보유 해당코인 가치
    }
    portfolio_analysis.append(my_portfolio)
    total_value += total_price

  for item in portfolio_analysis:
    item["weight"] = (item["total_price"] / total_value) * 100

  return portfolio_analysis, total_value

portfolio_analysis, total_value = get_current_prices_api()

template = """
코인명: {}
보유수량: {}
현재가: {}
보유가치: {}
비중: {}
"""

for item in portfolio_analysis:
  message = template.format(item['coin'], item['count'], item['trade_price'], item['total_price'], item['weight'])
  print(message)
```

```python
from datetime import datetime
import time
import requests

# 가격알림 시스템
# 목표가 설정, 현재가 모니터링, 목표가 도달 시 알림메시지 출력

alert_count = 0  # 알람 울린 횟수

# 코드 작성당시 가치로 설정함
high_target = 215
low_target = 210

def get_single_price_api():

  url = "https://api.upbit.com/v1/ticker?markets=KRW-DOGE"
  headers = {"accept": "application/json"}
  response = requests.get(url, headers=headers)
  json_resp = response.json()
  trade_price = json_resp[0]["trade_price"]
  current_time = datetime.now().strftime("%H:%M:%S")

  if (trade_price >= high_target):
      alert_count += 1
      print(f"설정한 목표상한가 {high_target}에 도달했습니다.")
  elif (trade_price <= low_target):
      alert_count += 1
      print(f"설정한 목표하한가 {low_target}에 도달했습니다.")
  else :
    print(f"[{current_time}] 현재 가격은 {trade_price} 입니다")


while alert_count <= 3:
  get_single_price_api()
  time.sleep(10)
```

```python
# 암호화폐 수익률 계산기
# 과거 특정 시점에 투자했다면 현재 수익률이 얼마인지 계산하는 프로그램을 작성하시오:
# 1. 투자 시점과 투자 금액 설정
# 2. 해당 시점의 가격과 현재 가격 비교
# 3. 수익률과 수익금 계산

import requests

market = "KRW-DOGE"
to = "2026-01-01 02:00:00" # 투자날짜
invest_price = 2000000 # 투자금액

def get_historical_data_api():
  converting_price_unit = "KRW"
  params = {
      "market":market,
      "to": to,
      "convertingPriceUnit": converting_price_unit
  }

  url = "https://api.upbit.com/v1/candles/days"
  headers = {"accept": "application/json"}
  response = requests.get(url, headers=headers, params=params)

  json_resp = response.json()
  return json_resp[0]['trade_price']

def get_single_price_api():
  url = "https://api.upbit.com/v1/ticker"
  headers = {"accept": "application/json"}
  params = {
      "markets":market
  }

  response = requests.get(url, headers=headers, params=params)
  json_resp = response.json()
  trade_price = json_resp[0]["trade_price"]

  return trade_price

past_doge_price = get_historical_data_api()

cur_doge_price = get_single_price_api()

def calculate_investment_return():
  doge_count = invest_price / past_doge_price
  cur_value = doge_count * cur_doge_price

  my_return = (cur_value - invest_price)
  my_return_rate = my_return / invest_price * 100

  return doge_count, cur_value, my_return, my_return_rate

doge_count, cur_value, my_return, my_return_rate = calculate_investment_return()

template = """
투자일자: {}
투자금액: {}
투자 시점 가격: {}
구매 수량: {}
현재 가격: {}
현재 가치: {}
손익: {}
수익률: {}
"""

message = template.format(to,invest_price,past_doge_price,doge_count,cur_doge_price,cur_value,my_return,my_return_rate)
print(message)
```
